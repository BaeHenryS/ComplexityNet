```python
def map_values(val):
    if val in [1,2]:
        return 1
    elif val in [3,4]:
        return 2
    elif val in [5]:
        return 3

df = pd.read_csv('model2.csv')
# Compare the accuracies by comparing the "Actual" column value and "Fine-Tuned" column value:
print("accuracy: ", (df['Actual'] == df['Fine-Tuned']).sum() / len(df))

df['Original'] = df['Original'].map(map_values)
df['Fine-Tuned'] = df['Fine-Tuned'].map(map_values)
df['Actual'] = df['Actual'].map(map_values)
df = df.fillna(0)
```

```python
# Compare the accuracies by comparing the "Actual" column value and "Fine-Tuned" column value:

df['Actual'] = df['Actual'].astype(int)
df['Original'] = df['Original'].astype(int)
df['Fine-Tuned'] = df['Fine-Tuned'].astype(int)

# Calculate the accuracy:

accuracy_finetune = (df['Actual'] == df['Fine-Tuned']).sum() / df.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_finetune * 100))

accuracy_original = (df['Actual'] == df['Original']).sum() / df.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_original * 100))
```

```python
accuracies = [61.76, 50.0, 52.94 ]
accuracies = np.divide(accuracies, 100)

og_accuracy = [35.29, 32.35, 29.41]
avg_og = np.mean(og_accuracy)
avg_og = np.divide(avg_og, 100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define data and colors
models = ['Original', 'Model 1', 'Model 2', 'Model 3']
colors = ['green', 'blue', 'blue', 'blue']
barplot_data = [avg_og, accuracies[0], accuracies[1], accuracies[2]]

# Create a DataFrame for Seaborn plotting
df = pd.DataFrame(list(zip(models, barplot_data)), columns=['Method 1 Models', 'Accuracy'])

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Create bar plot with increased resolution
plt.figure(figsize=(6, 4), dpi=300)  # Increase resolution to 300 dpi
bars = sns.barplot(x='Method 1 Models', y='Accuracy', data=df, palette=colors, width=0.4)  # Make bars narrower

# Add numbers on top of each bar
for idx, row in df.iterrows():
    bars.text(idx, row.Accuracy + 0.02, round(row.Accuracy, 2), color='black', ha="center", weight='bold')

plt.ylim(0, 1)

# Make y-axis label bold
plt.ylabel('Accuracy', labelpad=10, weight='bold')

# Make title bold
plt.title('Accuracy of Complexity Models', weight='bold')

# Make x-axis label bold
plt.xlabel('1 Trial Labeling Method', labelpad=10, weight='bold')

plt.show()
```

```python
# Import Model4.csv

df4 = pd.read_csv('model4.csv')

df4.head(60)
```

```python
# First column is the actual value, 2-6 columns are fine tuned values, 7-11 columns are original values

# Create a column called finetune_common, which is the most common value in the fine tuned columns
df4['finetune_common'] = df4.iloc[:, 1:6].mode(axis=1)[0]

# Create a column called original_common, which is the most common value in the original columns
df4['original_common'] = df4.iloc[:, 6:11].mode(axis=1)[0]

# Create a column called finetune_mean, which is the mean of the fine tuned columns
df4['finetune_mean'] = df4.iloc[:, 1:6].mean(axis=1)

# Create a column called finetune_mean_int, which is the mean of the fine tuned columns rounded to the nearest integer
df4['finetune_mean_int'] = df4['finetune_mean'].round()

# Create a column called original_mean, which is the mean of the original columns
df4['original_mean'] = df4.iloc[:, 6:11].mean(axis=1)

# Create a column called original_mean_int, which is the mean of the original columns rounded to the nearest integer
df4['original_mean_int'] = df4['original_mean'].round()

# Create a column called finetune_std, which is the standard deviation of the fine tuned columns
df4['finetune_std'] = df4.iloc[:, 1:6].std(axis=1)

# Create a column called original_std, which is the standard deviation of the original columns
df4['original_std'] = df4.iloc[:, 6:11].std(axis=1)

 
```

```python
# Compare the accuracies by comparing the "Actual" column value and "Fine-Tuned" column value:

df4['Actual'] = df4['Actual'].astype(int)
df4['finetune_common'] = df4['finetune_common'].astype(int)
df4['original_common'] = df4['original_common'].astype(int)

# Calculate the accuracy:


accuracy_finetune = (df4['Actual'] == df4['finetune_common']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_finetune * 100))

accuracy_original = (df4['Actual'] == df4['original_common']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_original * 100))
```

```python
# Do the same for the mean values:

accuracy_finetune = (df4['Actual'] == df4['finetune_mean_int']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_finetune * 100))

accuracy_original = (df4['Actual'] == df4['original_mean_int']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_original * 100))
```

```python
def map_values(val):
    if val in [1,2]:
        return 1
    elif val in [3,4]:
        return 2
    elif val in [5]:
        return 3
    

# Convert the values in the "Actual" column and finetune_common, original column to 1, 2, 3
df4['Actual'] = df4['Actual'].map(map_values)
df4['finetune_common'] = df4['finetune_common'].map(map_values)
df4['original_common'] = df4['original_common'].map(map_values)

# Calculate the accuracy:
accuracy_finetune = (df4['Actual'] == df4['finetune_common']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_finetune * 100))

accuracy_original = (df4['Actual'] == df4['original_common']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_original * 100))
```

```python
# Do the Mapping for the mean values:

df4['finetune_mean_int'] = df4['finetune_mean_int'].map(map_values)
df4['original_mean_int'] = df4['original_mean_int'].map(map_values)

# Calculate the accuracy:
accuracy_finetune = (df4['Actual'] == df4['finetune_mean_int']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_finetune * 100))

accuracy_original = (df4['Actual'] == df4['original_mean_int']).sum() / df4.shape[0]
print("Accuracy: {:.2f}%".format(accuracy_original * 100))


accuracy_finetune_model4 = accuracy_finetune
accuracy_original_model4 = accuracy_original


sd_finetune_model4 = df4['finetune_std'].mean()
sd_original_model4 = df4['original_std'].mean()
```
